# numpy

## Cách tạo một số `ndarray` đặc biệt

`numpy` cung cấp một số hàm để tạo nhanh một số dạng `ndarray`, một số hàm thường dùng nhu sau.

|Hàm|Mô tả|
|-|-|
|`np.arange(start, stop, step)`|Tạo một `array` gồm các phần từ cách đều nhau, các phần tử cách nhau một khoảng là `step`|
|`np.linspace(start, stop, num)`|Tạo một `array` gồm các phần từ cách đều nhau có `num` phần tử|
|`np.empty(shape)`|Tạo một `array` trống với kích thước `shape`|
|`np.zeros(shape)`|Tạo một `array` với các phần tử bằng 0 với kích thước `shape`|
|`np.ones(shape)`|Tạo một `array` với các phần tử bằng 1 với kích thước `shape`|



```python
import numpy as np

# Lưu ý về np.arange(start, stop, step):
# - Có thể bỏ qua start, step.
# - array không bao gồm giá trị stop

print('Dạng đầy đủ:', np.arange(0, 10, 2))
print('Không có start, step:', np.arange(10))
```

```python
# Lưu ý về np.linspace(start, stop, num)
# - Giá trị mặc định của num là 50.
# - array kết quả có bào gồm giá trị stop

print(np.linspace(0, 1, 11))
```

```python
shape = (2, 3)

# tạo array trống có 2 dòng 3 cột
print(np.empty(shape))
```

```python
print(np.zeros(shape))
```

```python
print(np.ones(shape))
```

Ngoài ra, có thể dùng các hàm `np.empty_like(arr)`, `np.zeros_like(arr)` và `np.ones_like(arr)` để tạo nhanh các dạng `array` cùng kích thước với tham số `arr`.

```python
arr = np.array([[1, 2, -1], [3, 4, 5]])
print(np.zeros_like(arr))
```

## Một số hàm tính toán khác

Bên cạnh các phép toán cơ bản (`+`, `-`, '*', `/`), `numpy` còn cung cấp nhiều hàm để tính toán như tron [trang tài liệu](https://numpy.org/devdocs/reference/routines.math.html).

```python
# kiểm tra đẳng thức sin²θ + cos²θ = 1
a = np.pi * np.linspace(0, 1, 11)
print(np.sin(a) ** 2 + np.cos(a) ** 2)
```

# Matplotlib

## Giới thiệu

`matplotlib` là một package của python cung cấp các công cụ để vẽ đồ thị. Trong phần này, chúng ta tìm hiểu cách sử dụng `matplotlib` trong các trường hợp phổ biến.

```python
# import matplotlib
import matplotlib.pyplot as plt

# magic command
%matplotlib inline
```

## Biểu đồ đơn giản

Trong phần này, ta sẽ tìm hiểu cách vẽ một đồ thị đơn giản và cách thay đổi những thành phần có trong biểu đồ này.

```python
# đồ thị đơn giản
theta = np.linspace(-np.pi, np.pi, 100)

S = np.sin(theta)
C = np.cos(theta)

plt.plot(theta, S)
plt.plot(theta, C);
# dấu ; để tắt hiển thị kết quả của dòng cuối cùng khi thực thi ô
```

### Điều chỉnh thuộc tính nét vẽ

`plt.plot()` có một số tham số để điều chỉnh thuộc tính của nét vẽ như:

|tham số|mô tả|
|-|-|
|color|màu nét vẽ|
|linewidth|độ dày nét|
|linestyle|kiểu nét (`-`, `--`, `-.`, ...)|
|...|

Chi tiết có thể xem thêm tại [trang tài liệu](https://matplotlib.org/2.0.2/users/pyplot_tutorial.html#controlling-line-properties).

```python
# tạo biểu đồ
plt.figure(figsize=(10, 6), dpi=80)
plt.plot(theta, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(theta, S, color="red",  linewidth=2.5, linestyle="-.");

# lưu biểu đồ để cập nhật
fig_to_update = plt.gcf()
```

### Điều chỉnh giới hạn biểu đồ
Đề điều chỉnh giá trị lớn nhất, nhỏ nhất cho trục tọa độ, có thể dùng:
- `plt.xlim(min, max)` cho trục hoành (trục Ox).
- `plt.ylim(min, max)` cho trục hoành (trục Oy).

```python
# lấy biểu đồ cũ để cập nhật thuộc tính
plt.figure(fig_to_update)

# chỉnh giới hạn trục tọa độ
plt.xlim(-4, 4)
plt.ylim(-1.2, 1.2)

# hiển thị
fig_to_update
```

### Thêm trục tọa độ
Để thêm trục tọa độ một cách đơn giản, ta có thể dùng:
- `plt.axhline()`
- `plt.axvline()`


```python
plt.figure(fig_to_update)

plt.axhline(0, color='black')
plt.axvline(0, color='black')

fig_to_update
```

Ngoài ra, có thể dùng đến `ax.spines()`, `ax.get_xticks()`, `ax.get_yticks` để có những tùy chỉnh cao hơn. Có thể xem trong [đường dẫn](https://python-course.eu/numerical-programming/spines-and-ticks-in-matplotlib.php).

### Thêm chú thích
Để thêm trục tọa độ một cách đơn giản, ta dùng thuộc tính `label` của hàm `plt.plot()`. Đề điều chỉnh vị trí của chú thích, có thể dùng `plt.legend()` với tham số `loc`.

```python
from matplotlib import legend
plt.figure(fig_to_update)

plt.plot(theta, C, color="blue", linewidth=2.5, linestyle="-", label = 'cos')
plt.plot(theta, S, color="red",  linewidth=2.5, linestyle="-.", label = 'sin');

plt.legend(loc = 'upper right') # góc trên bên phải
fig_to_update
```

### Plot styles
`plt` còn cung cấp một số mẫu định dạng đồ thị, có thể sử dụng thông qua `plt.style.use(<tên styles>)`. Các `<tên styles>` có thể dung được liệt kê tại [đây](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html). _Lưu ý_: styles cần được dùng khi import matplotlib.

```python
plt.style.use('ggplot')
plt.figure(fig_to_update)
fig_to_update
```

### Đóng đồ thị
Trước khi chuyển sang vẽ đồ thị khác, cần đóng lại đồ thị cũ bằng `plt.close()`.

```python
plt.close(fig_to_update)
```

## Biểu đồ con

`plt` cung cấp khả năng vẽ nhiều đồ thị trên nhiều khung khác nhau bằng cách sử dụng `plt.subplot()`

```python
# ví dụ vẽ hai đồ thị theo chiều dọc
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)


t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), color='tab:blue', marker='o')
plt.plot(t2, f(t2), color='black')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), color='tab:orange', linestyle='--');
```

```python
plt.close()
```

Các bước vẽ nhiều đồ thị:
1. Dùng `plt.figure()` để tạo hình vẽ.
2. Sử dụng `plt.subplot()` để chuyển sang vùng thích hợp.
3. Vẽ đồ thị
4. Đóng đồ thị

`plt.subplot()` có thể khai báo theo hai cách:
- `plt.subplot(a, b, c)`
- `plt.subplot(abc)`

Hai cách đều hiểu là sử dụng vùng tại vị trí `c` trong một bản vẽ được chia thành `a` dòng và `b` cột. Vị trí được đánh số từ 1 và theo thứ tự từ trái qua phải, từ trên xuống dưới.

```python
# bước 1, tạo bản vẽ
plt.figure()

theta = np.linspace(-np.pi, np.pi, 100)
# chia bản vẽ thành 3 dòng 1 cột
# dùng vùng 1
plt.subplot(3, 1, 1)
S = np.sin(theta)
plt.plot(theta, S, color = 'blue')

# dùng vùng 2
plt.subplot(3, 1, 2)
C = np.cos(theta)
plt.plot(theta, C, color = 'red', linestyle = '--')

# dùng vùng 3
plt.subplot(3, 1, 3)
T = np.tan(theta)
plt.plot(theta, T, color = 'orange', linestyle = '-.')
plt.ylim(-1.2, 1.2);
```

```python
plt.close('all')
```

## Lưu đồ thị

Ta có thể lưu đồ thị bằng cách sử dụng hàm `plt.savefig(<tên_tập_tin>)`. Trong đó, `<tên_tập_tin>` có thể là:
- Một tên duy nhất (như `'abc.png'`). Khi đó, hình ảnh sẽ được lưu cùng vị trí với vị trí tập tin `ipynb` này.
- Có thêm vị trí thư mục (như `'D:\fig\abc.jpg'`). Khi đó, hình ảnh sẽ được lưu tại thư mục chỉ định.

_Lưu ý_: `plt.savefig()` cần được sử dụng trước `plt.close()`.

```python
# vẽ đồ thị và lưu thành tập tin fig.png
r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

fig, ax = plt.subplots(1, 1, figsize=(5, 8), subplot_kw={'projection': 'polar'},
                        layout='constrained')
ax.plot(theta, r)
ax.set_rmax(2)
ax.grid(True)
plt.savefig('fig.png') # <------ Lưu hình
plt.close()
```
